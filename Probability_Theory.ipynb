uncertainty is the key of pattern recognition. to calculate it we need to know Probability.
Probability: the probability of an event to be the fraction of times that event occurs out of the total number of trials.

Probability = P(event) / p(total number of trials).

sum rule p(X) = summation(p(X, Y )) 
product rule p(X, Y ) = p(X|Y)p(Y).

# Example 1

Imagine we have two boxes, one red and one blue, and in the red box
we have 2 apples and 6 oranges, and in the blue box we have 3 apples and 1 orange.
This is illustrated in Figure 1.9. Now suppose we randomly pick one of the boxes
and from that box we randomly select an item of fruit, and having observed which
sort of fruit it is we replace it in the box from which it came. We could imagine
repeating this process many times. Let us suppose that in so doing we pick the red
box 40% of the time and we pick the blue box 60% of the time, and that when we
remove an item of fruit from a box we are equally likely to select any of the pieces
of fruit in the box.
image.png

# 1- what is the overall probability that the selection procedure will pick an apple?
# 2- given that we have chosen an orange


```python
blue_box = ['orange', 'apple', 'apple', 'apple']
red_box = ['orange', 'orange', 'orange', 'orange', 'orange', 'orange', 'apple', 'apple']

total_red_box = len(red_box)
total_blue_box = len(blue_box)

prob_red_box = 0.4
prob_blue_box = 0.6

# Answer of question 1
# To calculate the probability of the selection procedure will pick an apple we need to use sum and product rules
# sum rule p(apple) = p(apple, blue_box ) + p(apple, red_box )
# p(apple, blue_box ) = p(apple|blue_box).p(blue_box)
# p(apple, red_box ) = p(apple|red_box).p(red_box)
```

```python
from collections import Counter
# With sequence of items 
red_box_sample_space = Counter(red_box)
# with keyword arguments
blue_box_sample_space = Counter(blue_box)
```

```python
apple_prob = (red_box_sample_space['apple'] / total_red_box) * prob_red_box + (blue_box_sample_space['apple'] / total_blue_box) * prob_blue_box
```

```python
print("the probability of the selection procedure will pick an apple =", apple_prob)
```

```python
orange_prob = 1 - apple_prob
print("the probability of the selection procedure will pick an orange =", orange_prob)
```

# Suppose instead we are told that a piece of fruit has been selected and it is an orange, and we would like to know which box it came from.Suppose instead we are told that a piece of fruit has been selected and it is an orange, and we would like to know which box it came from.

```python
# to solve this question we need to use Bayes’ theorem
# p(blue_box | orange) = p(orange | blue_box).P(blue_box) / p(orange)
# p(red_box | orange) = p(orange | red_box).P(red_box) / p(orange)

prob_blue_given_orange = ((blue_box_sample_space['orange'] / total_blue_box) * prob_blue_box) / orange_prob

prob_red_given_orange = 1 - prob_blue_given_orange
```

```python
print("Probability of blue box given orange =", prob_blue_given_orange)
print("Probability of red box given orange =", prob_red_given_orange)
```
